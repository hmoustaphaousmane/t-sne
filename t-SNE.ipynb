```python
# Importer les bibliothèques nécessaires :
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
```

```python
# Charger les données de Iris (à partir de la bibliothèque `sklearn.datasets`)
iris = datasets.load_iris ()

# Récuperer les données dans x, et les étiquettes dans y.
x = iris.data
y = iris.target
```

```python
# Réduire la dimensionnalité (à 2 dimensions) avec t-SNE : n_components = 2 par défaut
tsne = TSNE ()

# Ajuster x dans un espace intégré et récupérer la sortie transformée dans x_tsne
x_tsne = tsne.fit_transform (x)
```

```python
# Visualiser les résultats
plt.figure (figsize= (10, 6))
colors = ['red', 'green', 'blue']
for i in range (3):
    plt.scatter (x_tsne [y == i, 0], x_tsne [y == i, 1], c = colors [i], label = iris.target_names [i])
plt.title ("t-SNE sur les données Iris")
plt.legend (loc = 'best')
plt.xlabel ("Dimension 1")
plt.ylabel ("Dimension 2")
plt.show ()
```
